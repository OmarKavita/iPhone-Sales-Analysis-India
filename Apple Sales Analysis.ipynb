Now, we will try to find the top rated products and draw visualization graph comparing the rating of all the products.

```python
Top_Rating = df.sort_values(['Star Rating'], ascending=False).head(10)
Top_Rating
```

These are ten top rated products. Visualizing it on histogram:

```python
sns.countplot(x='Star Rating', data=df, palette='flare');
```

Most of the Iphones have got 4.6 rating out of 5.

Now let's have a look at number of ratings of highest rated iPhones on flipkart:

```python
sns.barplot(x='Product Name', y='Number Of Ratings', data=Top_Rating, palette='gist_rainbow')
plt.xlabel('Product')
plt.ylabel('Rating Count')
plt.title('Number of ratings for highest rated iPhones')
plt.xticks(rotation=45)
plt.show()
```

According to the above bar graph, Apple iphone 8 plus(Gold, 64 GB) has the most number of ratings on flipkart.

Now let's have a look at the number of reviews:

```python
sns.barplot(x='Product Name', y='Number Of Reviews', data=Top_Rating, palette='gist_rainbow')
plt.xlabel('Products')
plt.ylabel('Reviews Count')
plt.title('Number of reviews of highest rated iPhones')
plt.xticks(rotation=45)
plt.show()
```

we can see here that Apple iPhone 8 plus(Gold, 64 GB) also has the most numbers of reviews on flipkart. 

Now let us find the relationship between sale price of iPhone and raings:

```python
sns.regplot(x='Number Of Ratings', y='Sale Price', data=df, color='purple')
plt.xlabel('Sale Price')
plt.ylabel('Number Of Ratings')
plt.title('Relaionship between Sale price and Number of ratings')
plt.show()
```

There is a negative linear relationship between the sale price of iPhones and the number of ratings. It means iPhones with lower sale prices are sold more in India. 

Now let’s have a look at the relationship between the discount percentage on iPhones on Flipkart and the number of ratings:

```python
sns.regplot(x='Number Of Ratings', y='Discount Percentage', data=df, color='purple')
plt.xlabel('Number of Ratings')
plt.ylabel('Discount(%)')
plt.title('Relationship between Discount Percentage and Number of Ratings')
plt.show()
```

There is a linear relationship between the discount percentage on iPhones on Flipkart and the number of ratings. It means iPhones with high discounts are sold more in India.
